# Polynomial Regression Assignment

```python
import numpy as np
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, preprocessing
import numpy.polynomial.polynomial as poly
%matplotlib inline
```

# Importing Data 
Import data from poly_regression.csv

```python
# Run the code cell
df = pd.read_csv('poly_regression.csv')
data = df.values
X = data[:,0]
Y = data[:,1]
X = X.reshape(X.shape[0],1)
Y = Y.reshape(Y.shape[0],1)
```

# Training and Test Set
Divide the dataset approximately into 2 equal parts : training and test set. 

```python
nX_tr = X.shape[0] // 2
X_tr = X[:nX_tr]
X_test = X[nX_tr:]

nY_tr = Y.shape[0] // 2
Y_tr = Y[:nY_tr]
Y_test = Y[nY_tr:]
```

# Plotting Dataset

```python
# Write code to plot the dataset.
plt.scatter(X_tr, Y_tr, c='r')
plt.scatter(X_test, Y_test, c='b')
plt.show()
```

# Model Selection.
Write code to obtain polynomial regression models of different orders starting from linear regression i.e. degree = 1 to higher degree models like degree = 2 to 10. Find training and test error for every order and plot these errors v/s degree. Select the order that fits the data best based on low training and test error. You can use poly.polyval method from numpy to find coefficients of the different models.

```python
# initialize list of tested degrees
degrees = np.array(range(1, 10))
MSEtr = []
MSEtest = []

for d in degrees :  # test each degree
    w_hat = poly.polyfit(X_tr.flatten(), Y_tr.flatten(), d)
    y_hat = poly.polyval(X_tr, w_hat)
    MSEd = np.mean((y_hat - Y_tr) ** 2)
    MSEtr.append(MSEd)
    
    y_hat = poly.polyval(X_test, w_hat)
    MSEd = np.mean((y_hat - Y_test)**2)
    MSEtest.append(MSEd)
    
plt.plot(degrees, MSEtr, 'bo-')
plt.plot(degrees, MSEtest, 'ro-')
plt.xlabel('Model order')
plt.ylabel('MSE')
plt.legend(['Training', 'Test'])
plt.show()

opt_degree = np.argmin(MSEtest)
opt_degree_tr = np.argmin(MSEtr)

print("Optimal degree is:", opt_degree)
print("Because MSEtest is minimized at d = ", opt_degree)
```

```python
plt.scatter(X, Y)
w = poly.polyfit(X.flatten(), Y.flatten(), opt_degree)

yhat = poly.polyval(np.sort(X.flatten()), w)

plt.plot(np.sort(X.flatten()), yhat, 'r-', linewidth=3)
plt.title("Data vs. Fitted Polynomial Regression")
plt.legend(["Regression", "Data"])
plt.show()
```
